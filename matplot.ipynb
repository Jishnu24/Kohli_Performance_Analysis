```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.legend(['sine','cosine'])
plt.show()
```

```python
#create the first subplot
plt.subplot(2,1,1)
plt.plot(x,y_sin)
#create the second subplot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
#plt.xlabel('x axis label')
#plt.ylabel('y axis label')
#plt.legend(['sine','cosine'])
plt.show()
```

```python
#create the first subplot
plt.subplot(3,1,1)
plt.plot(x,y_sin)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.legend(['sine'])
#create the second subplot
plt.subplot(3,1,3)
plt.plot(x,y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.legend(['cosine'])
plt.show()
```
